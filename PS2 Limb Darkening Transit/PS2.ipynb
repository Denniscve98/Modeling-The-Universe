```python
# Problem Set #2
# Dennis Cruz dec2174
# Collaborators: Lori Elizabeth Porter, lep2176


# Problem 1: Higher Order Integrator

import math as m
import numpy as np
from numpy import sin, pi

# Bounds:
# lower_b = 0
# upper_b = pi/2.0

def function(x): #function to be integrated
    return sin(x)

def trap_integ (function, N, upper_b, lower_b):                         # Trapezoidal rule for integration
    delta_x   = (upper_b - lower_b) / float(N)                          # N is the number of steps
    area_trap = delta_x * 0.5 * (function(lower_b) + function(upper_b)) # Area of a trapezoid
    for i in range (0, N):
        area_trap = area_trap + delta_x * function (i * delta_x) 
    return area_trap

# Showing results at different Ns:
j = 1 
while j < 6:
    print ("f(x) =", trap_integ(function, 10**j, pi/2.0, 0))          # Evaluating over N up to 10**5
    j += 1

# Calculating fractional error:

j = 1
while j < 6:                                                          # Loop that goes from 1 to 5 to evaluate N up to 10**5
    true_value     = 1                                                # True value of sin (x) from 0 to pi/2
    estimate_value = trap_integ (function, 10**j, pi/2.0, 0)          # Estimate at N = 10**5
    fract_error    = m.fabs(true_value - estimate_value) / float(true_value)
    print ("Fractional error at", (10**j), "=" , fract_error)
    j += 1
    
# Fractional error is much lower than using the Rectangular rule. Order 10e-3
```

```python
# Problem 2
# Limb Darkening

import math as m
import numpy as np

# p = 0.2 #ratio of the planet radius to the stellar radius
# z = 0.9 #separation of the centers

# Bounds of integration for I(r):
# lower_b = 0
# upper_b = 1

def I (r): # Defining I(r) with no limb- darkening = 1
    return 1

def delta (p, r, z): #function for different conditions of delta
    
    if r >= (z + p) or r <= (z - p):
        result = 0.0
        return result
        
    elif (r + z) <= p:
        result = 1
        return result
        
    else:
        k_value = (z**2 - p**2 + r**2) / (2.0*z*r) # Value to evaluate acos
        if -1 <= k_value <= 1:                     # Boundaries of acos to avoid math error
            k = m.acos (k_value)                   # "k" variable substitutes the equation
            result = (1/m.pi) * k
        return result

# Testing delta function for different "r":
for r in np.arange (0, 1.5, 0.5):
    print (r, delta(0.2, r, 0.9))
# It works!

# Analytical solution to F(p, |z|):
def ana_solution (upper_b, lower_b):
    return (I(r) * (1 - (delta(0.2, r, 0.9))) * ((1/2.0) * (2*upper_b**2 - 2*lower_b**2))) / (I(r) * ((1/2.0) * (2*upper_b**2 - 2*lower_b**2)))

print ('\n')
print ("Analytical Solution: F(p, |z|) =", ana_solution (1, 0))
```
