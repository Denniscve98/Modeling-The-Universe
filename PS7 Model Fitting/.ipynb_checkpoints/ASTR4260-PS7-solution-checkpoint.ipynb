```python
import math
import numpy as np
import matplotlib.pyplot as plt

import lightkurve as lk
import emcee               #we'll need this for the graduate problem

def get_lightcurve(name):
    lcs      = lk.search_lightcurve(name, mission="Kepler", cadence="short").download_all()
    time     = lcs[0].time.value                    # get the time of each observation
    flux     = np.array(lcs[0].flux.value.data)     # get the flux
    flux_err = np.array(lcs[0].flux_err.value.data) # get the error in the flux
    return (time, flux, flux_err)
```

Next, extract the section from t = 261 to 262 and apply the two-sigma clipping algorithm to get an estimate of the continuum flux.

```python
# Extract and normalize a section of the data
def ComputeClippedMean(x):
    for i in range(8):
        x = x[np.where(abs(x - x.mean()) < 2*x.std())]
    return x.mean()


def prep_data(t0, t1, time, flux, flux_err):
    index     = np.where((time > t0) & (time < t1) & np.isfinite(flux_err))   #note the test to eliminate NaN's and inf's
    time_cut  = time    [index]  
    flux_cut  = flux    [index]
    error_cut = flux_err[index]
    
    FluxNormalization = ComputeClippedMean(flux_cut)
    print('Flux normalization = ',FluxNormalization)    #check this makes sense
    FluxRatioObserved = flux_cut/FluxNormalization
    FluxError = error_cut/FluxNormalization
    return (time_cut, FluxRatioObserved, FluxError)
```

Normalize the flux and the error

```python
t0 = 261; t1 = 262
time, FluxRatioObserved, FluxError = prep_data(t0, t1, time, flux, flux_err)
```

Now, turn to computing the model curve.  First, we'll define the geometrical overlap function $\delta$, the limb-darkening function and the two functions (numerator and denominator) that we need to integrate, as in Problem Set \#2. (Feel free to consult the solution notebook.)

```python
def delta_arr(p, r, z):
    "returns delta for all values of an array r, with scalars p & z"
    
    #at the ends of Simpson's rule, scalar values are fed in
    if np.isscalar(r):
        nr = 1
        r  = np.array([r])
    else:
        nr = len(r)
        
    result = np.zeros(nr) - 99                        #mark unmodified elements
    
    #where returns an array of indices satisfying the condition
    result[ np.where(r + z <= p)]                                              = 1.0

    #two conditions tested, and concatenated inside a list
    result[np.concatenate((np.where(r >= z + p)[0], np.where(r <= z - p)[0]))] = 0.0  
    
    # zero protect the division by adding a very small number to it
    eps = 1e-14
    
    #fill everything else
    ind = np.where(result == -99)[0]
    result[ind] = np.arccos((z*z - p*p + r[ind]**2) / (2.0 * z * r[ind] + eps)) / np.pi
    
    return result
```

```python
def I(r):
    '''A Limb-darkening function'''
    mu = (1 - (r**2))**(0.5)
    return 1 - (1 - (mu)**0.5)

def numerator(r, p, z):
    return I(r) * (1 - delta_arr(p,r,abs(z))) * 2 * r

def denominator(r, p, z):
    return I(r) * 2 * r
```

Finally, define the function to do the integration necessary to compute the predicted transit light curve given the three parameters (p, t0, tau).  We'll use scipy's Gaussian quadrature to do the integration as this is very efficient.

```python
from scipy.integrate import quad

def ComputeLightCurve(time, p, t0, tau, eps=1e-3, plot=False):
    """ Convert from Time to z and compute predicted transit curve """
    zs = (time-t0)/tau
    FluxRatio = []
    for zval in zs:
        FluxObscured, FO_error = quad(numerator, 0.0, 1.0, epsrel=eps, args=(p,zval))
        FluxUnobscured, FU_error = quad(denominator, 0.0, 1.0, epsrel=eps, args=(p,zval))
        FluxRatio.append(FluxObscured/FluxUnobscured)
    if (plot): 
        plt.plot(zs,FluxRatio)  #check that everything is OK
        plt.xlabel('z'); plt.ylabel('Flux Ratio')
    return np.array(FluxRatio)
```

Now that everything is set up, it's easy to compute the profile.

```python
print(len(time), time.min(), time.max())
```

```python
p = 0.0775; tau = 0.1; t0 = 261.3

FluxRatioTheory = ComputeLightCurve(time, p, t0, tau, eps = 1e-4, plot = True)
```

And we can compute the chi-squared value for this.

```python
def ComputeChiSquared(yobs, ymodel, ysigma):
    return np.sum(((yobs-ymodel)/ysigma)**2)

chi2 = ComputeChiSquared(FluxRatioObserved, FluxRatioTheory, FluxError)
print("chi-squared = ", chi2)
```

### Problem 2: Plot and then check probability of this fit

First, let's plot the data and the fit

```python
def comparison(time, FluxRatioObserved, FluxError, FluxRatioTheory):
    #zorder determines which plot is on top
    plt.errorbar(time, FluxRatioObserved, yerr=FluxError, fmt='g.', zorder=1)
    plt.plot(time, FluxRatioTheory,'k-',linewidth=2, zorder=2)

    plt.xlabel("Coordinated Universal Time - 2454833 (days)")
    plt.ylabel("Flux Ratio (normalized)")
    plt.show()
    return

comparison(time, FluxRatioObserved, FluxError, FluxRatioTheory)
```

That doesn't look so bad to me!

Now, compute the probability.  Assuming Gaussian errors, we can use the incomplete gamma discussed in class to compute the likelihood of getting this chi-squared value.  We could integrate it ourselves, or just use scipy's version of the routine.  

```python
from scipy.special import gammaincc

def computeQ(nu, chi2):
    """Takes the number of degrees of freedom nu and a chi^2 value chi2"""
    print("degrees of freedom = ", nu)
    Q = gammaincc(nu/2, chi2/2)
    print("probability of this chi-squared = ",Q)
    return Q

Q = computeQ(len(FluxRatioObserved)-3, chi2)  #the fit has 3 parameters, and we have len(FluxRatioObserved) data points.
```

### Problem 3: Vary tau while keeping other two parameters constant

To do this, we can set up a range of tau values, and for each one, compute the theory curve and then determine the chi-squared value.  Plot this to see what it is like. (Even at delta tau = 0.005, this takes almost two minutes to run.)

```python
def vary_tau(time, FluxRatioObserved, FluxRatioTheory, FluxError, tau0 = 0.08, tau1 = 0.13, delta_tau=0.001):
    tau_list = np.arange(tau0, tau1, delta_tau)
    chi2_list = []
    for i in range(len(tau_list)):
        FluxRatioTheory = ComputeLightCurve(time, p, t0, tau_list[i])
        chi2_list.append(ComputeChiSquared(FluxRatioObserved, FluxRatioTheory, FluxError))

    #print (tau_list, np.array(chi2_list))
    plt.plot(tau_list, np.array(chi2_list))
    plt.xlabel("tau")
    plt.ylabel("chi-squared")
    plt.show()
    best_fit_index = np.argmin(chi2_list)
    return tau_list[best_fit_index]

d_tau = 0.005
tau_best_fit = vary_tau(time, FluxRatioObserved, FluxRatioTheory, FluxError, delta_tau = d_tau)
```

Repeat in a small region around best fit point and plot again, to better see the behavior of chi-squared right around the minimum. 

```python
tau_best_fit = vary_tau(time, FluxRatioObserved, FluxRatioTheory, FluxError, tau0 = tau_best_fit-d_tau, 
         tau1 = tau_best_fit+d_tau, delta_tau = d_tau/20.0)
```

Determine the best fit tau value by finding the minimum in the chi-squared

To determine the uncertainty in tau, we remember from class that for one parameter, one varies chi-squared by 1. So, we can find the range of tau values for which chi-squared < min chi-squared + 1.0

```python
def uncertainty(chi2_list, tau_list, tau_best_fit, best_fit_index):
    index = np.where(chi2_list <= chi2_list[best_fit_index] + 1.0)
    tau_range = tau_list[index[0][-1]] - tau_list[index[0][0]]
    print("Tau uncertainty = ", tau_range/2.0)
    return tau_range/2.0

tau_err = uncertainty(chi2_list, tau_list, tau_best_fit, best_fit_index)
```

This is better but still a bad fit, so uncertainty range doesn't actually mean anything!

### Problem 4 - Vary all three parameters

This is the hard one.  We vary all parameters at the same time.  There are lots of approaches to this: we could vary the parameters over a grid and find the minimum chi-squared.  However, another approach is the Monte Carlo technique -- there is a very nice tool called `emcee` -- let's use that.

To do so, we first need to define the -ln(Prob), which is basically chi-squared for Gaussian errors.

```python
#debug
time
!echo $PYTHONPATH
```

```python
#import emcee

def lnprob(x, y, yerr):
    p, t0, tau = x     
    FluxRatioTheory = ComputeLightCurve(time, p, t0, tau)
    return -0.5 * ComputeChiSquared(y, FluxRatioTheory, yerr)
```

```python
#debug
np.log(np.e)
```

Now define the number of "walkers" -- each walker is a chain of guesses which are accepted if they have a higher probability than the previous one, or they can also be accepted if lower with some probability.  We'll set up 10 walkers and initialize them in a small range around a guess (which I have set, slightly unfairly, to the best fit value determined from when I ran the procedure below the first time).

```python
ndim, nwalkers = 3, 10
p = 0.07719
tau = 0.10106
t0 = 261.50526
start_guess = np.array([p, t0, tau])
x0 = [start_guess + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
```

```python
x0
```

Now that everything is set up, call emcee with 500 samples for each walker.  It will return the sample chains.

```python
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(FluxRatioObserved, FluxError))
results = sampler.run_mcmc(x0, 500, progress = 'notebook')
```

That took over 24 hours to run!  (Note that adding the `progress` keyword to `run_mcmc` displayed a progress bar under the cell while it was running.)

Examine the returned sample chains.  Var = p, t_0, tau for 0, 1, 2.


```python
print(sampler.chain[0,0,:])
sampler.chain.shape
```

```python
import corner
```

```python
def show_walkers(sampler, labels, logs = False):
    fig, ax = plt.subplots(3,2,figsize = [10.5,8])
    plt.subplots_adjust(wspace=0.25)
    for var in range(3):
        #plt.subplot(3,2,var+1)
        for walker in range(10):
            ax[var][0].plot(np.abs(sampler.chain[walker,:,var]))
            ax[var][0].set_yscale('log')
            ax[var][0].set_ylabel(labels[var])
        #ax[var][1].show()
        #plt.subplot(3,2,(var+1)*2)
        for walker in range(10):
            ax[var][1].plot(sampler.chain[walker,:,var])
            ax[var][1].set_ylabel(labels[var])
        #ax[var][1].show()

    #set up a corner plot of the results
    burn_in = 80   # how many samples to skip for burn in
    if (not logs):
        samples = sampler.chain[:, burn_in:, :].reshape((-1, ndim))   #skip first 80 samples during burn-in
    else:
        samples = np.e**sampler.chain[:, burn_in:, :].reshape((-1, ndim))   #raise to the exponent
    fig = corner.corner(samples, show_titles=True, labels=["$p$", "$t_0$", "$\\tau$"], title_fmt = ".4f")
    return samples

samples = show_walkers(sampler, ('$p$','$t_0$','$\\tau$'))
```

These would look like reasonable results, except that planet size $p$ and eclipse length $\tau$ should both be positive definite!  Is the solution reasonable?  

```python
def check_params(time, FluxRatioObserved, FluxRatioTheory, FluxError,p, t0, tau):
    #p = 0.0477
    #tau = 0.1333
    #t0 = 261.4348
    FluxRatioTheory = ComputeLightCurve(time, p, t0, tau)
    chi2 = ComputeChiSquared(FluxRatioObserved, FluxRatioTheory, FluxError)
    print("chi-squared = ", chi2)
    print("probability of this chi-squared = ",gammaincc((len(FluxRatioObserved)-3)/2, chi2/2))
    comparison(time, FluxRatioObserved, FluxError, FluxRatioTheory)

check_params(time, FluxRatioObserved, FluxRatioTheory, FluxError, 0.0477, 261.4348, 0.1333)  #print corner plot most likely values
```

That doesn't look good!  What if we take the mean of the samples instead?

```python
#calculate best fit parameter and uncertainty
mean_vals = np.mean(samples, axis=0); std_vals = np.std(samples, axis=0)
print("The best fit parameter set for p,tau,t0 is ", mean_vals)
print("The uncertainty for p,tau,t0 is ",std_vals)
check_params(time, FluxRatioObserved, FluxRatioTheory, FluxError, mean_vals[0], mean_vals[1], mean_vals[2])
```

Still way off.  Instead, let's try using the logs of the quantities to avoid negative values, and see if that gives better results

```python
def lnprob_log(x, y, yerr):
    p, t0, tau = np.e**x     #to maintain these values positive definite
    FluxRatioTheory = ComputeLightCurve(time, p, t0, tau)
    return -0.5 * ComputeChiSquared(y, FluxRatioTheory, yerr)

ndim, nwalkers = 3, 10
p = 0.07719
tau = 0.10106
t0 = 261.3        #shifted this guess because eyeball says 261.5026 is much too large, and it pushes the walkers into the wrong regime.
start_guess_log = np.log(np.array([p, t0, tau]))   #to maintain these values positive definite
x0_log = [start_guess_log + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]

sampler_log = emcee.EnsembleSampler(nwalkers, ndim, lnprob_log, args=(FluxRatioObserved, FluxError))
results_log = sampler_log.run_mcmc(x0_log, 500, progress = 'notebook')
```

```python
samples_log = show_walkers(sampler_log, ('$p$','$t_0$','$\\tau$'), logs=True)

mean_vals_log = np.mean(samples_log, axis=0); std_vals = np.std(samples_log, axis=0)
print("The best fit parameter set for p,tau,t0 is ", mean_vals)
print("The uncertainty for p,tau,t0 is ",std_vals)
```

```python
check_params(time, FluxRatioObserved, FluxRatioTheory, FluxError, mean_vals_log[0], mean_vals_log[1], mean_vals_log[2])
```
