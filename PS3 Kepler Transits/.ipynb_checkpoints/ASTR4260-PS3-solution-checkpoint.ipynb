Execute the code given in the problem to import the Kepler data

```python
import lightkurve as lk
import numpy as np

import matplotlib.pyplot as plot
%matplotlib inline                  

# Make displayed plots larger
plot.rcParams['figure.figsize'] = [18, 11]
```

```python
def get_lightcurve(name):
    lcs      = lk.search_lightcurve(name, mission="Kepler", cadence="short").download_all()
    time     = lcs[0].time.value                    # get the time of each observation
    flux     = np.array(lcs[0].flux.value.data)     # get the flux
    flux_err = np.array(lcs[0].flux_err.value.data) # get the error in the flux
    return (time, flux, flux_err)
```

```python
(time, flux, flux_err) = get_lightcurve("Kepler-7b")
```

Check that the download was successful by checking the length of one of the arrays that we read in.

```python
print(len(flux))
```

That looks good!  Now we can plot the data.

```python
def plot_data(t, f, fe, ylabl = "Flux (counts/s)"):       #ylabl is an optional parameter with a default value
    """plot data f for time t with errorbars fe and ylabel ylabl"""
    plot.errorbar(t, f, yerr=fe)
    plot.xlabel('Coordinated Universal Time - 2454833 (days)')
    plot.ylabel(ylabl)
    plot.show()

plot_data(time, flux, flux_err)
```

### Problem 2

Let's examine one eclipse closely by extracting a bit of data around it.  There are lots of ways to do this, but one way is to estimate by eye where the first eclipse happens  and use numpy's where command to get the indices where the time array lies between these values.

```python
def eclipse(t_left, t_right, t, f, fe):
    """ t_left and t_right surround an eclipse in 
           an observation with time t, flux f, and flux error fe 
        Plot the region and return a dictionary containing data in that region
           with labels 't' = time, 'f' = flux, and 'fe' = flux_err """
    
    index = np.where((t > t_left) & (t < t_right))   #indexes inside eclipse
    time_obs = t[index]
    flux_obs = f[index]
    flux_err_obs = fe[index]

    plot_data(time_obs, flux_obs, flux_err_obs)
    data = {'t' : time_obs, 'f' : flux_obs, 'fe' : flux_err_obs}
    
    return data
```

Let's make a first guess.

```python
tleft = 260 ; tright = 262
first = eclipse(tleft, tright, time, flux, flux_err)
```

That caught the eclipse, but has a lot of background as well.  Let's focus in a bit more.

```python
tleft = 261.1 ; tright = 261.5
first = eclipse(tleft, tright, time, flux, flux_err)
```

Looks okay, but this plots the actual flux observed by Kepler, while we really want the ratio of the obscured flux to the unobscured flux. There are lots of ways to get estimates of the unobscured flux (away from the eclipse) -- later on we will talk about better ways, but here let's just assume that the first value in our range is a reasonable estimate of the unobscured flux and normalize by that.

```python
def normalize(flux_obs, flux_err_obs):
    flux_normalization = flux_obs[0]
    flux_normalized = flux_obs/flux_normalization
    flux_normalized_err = flux_err_obs/flux_normalization
    return {'fn' : flux_normalized, 'fen' : flux_normalized_err}

fnorm = normalize(first['f'], first['fe'])

#make sure that we did, in fact, normalize the data
assert fnorm['fn'].max() < 1.2 and fnorm['fn'].min() >= 0.0
```

To compute the flux, we'll need the flux computation from the first problem set; let's copy that in now.  The other option would have been to put this in a file and then import it.

```python
import math as m

error_code = -999   #for use in the following functions

def sqrt_factor(p, z):
    """calculation of square root factor for lam_e"""
    global error_code
    
    sfact1 = 4*z*z                #floating point errors caused
    sfact2 = (1 + z*z - p*p)**2   #sqrt to fail, so need to guard
    
    if (sfact1 > sfact2):         #NORMAL case
        sqrt_fact = m.sqrt(0.25*(sfact1 - sfact2))
    elif m.fabs((sfact2 - sfact1)/sfact1) < 1e-5:   #floating point zero!
        sqrt_fact = 0
    else:                #error trap: the sqrt should be >= 0
        print('z = ', z,' p = ',p, 'sfact1 = ', sfact1, 'sfact 2 = ', sfact2)
        sqrt_fact = error_code
            
    return sqrt_fact

def lam_e (p,z):
    """Obscuration function for planet traversing stellar disk
       Mandel & Agol (2002)
       p = planet radius normalized to stellar radius r*
       z = distance of planet center to stellar center normalized to r*"""
    global error_code

    # the first case is the planet off the star entirely.
    
    if z > 1.0 + p:      # reversing the inequality in the equation
        return 0.0
    
    # the next case can only occur if the "planet" is larger than 
    # the star, resulting in a total eclipse of the star
    
    elif z <= p - 1.0:   
        return 1.0
    
    # this case is the planet on the star entirely
    
    elif z <= 1.0 - p:   
        return p*p       
    
    # now deal with the edge case of partial obscuration
    
    elif (m.fabs(1.0-p) < z) and (z <= (1.0 +p)):
        
        sqrt_fact = sqrt_factor(p,z)
        
        arg_k0 = (p*p + z*z - 1.0) / (2.0*p*z)  
        arg_k1 = (1 - p*p + z*z) / (2.0*z)  
        
        kappa_0  = m.acos(arg_k0)
        kappa_1  = m.acos(arg_k1)
        
        #print(sqrt_fact, kappa_0, kappa_1)
        return (1.0/m.pi) * (p*p*kappa_0 + kappa_1 - sqrt_fact)
    
    else:                                  # error trap 
        print("unexpected value: z=", z)   
        return error_code                    


def FluxRatio(p, z):
    """ 
    Compute the ratio of obscured/unobscured flux for a planet transit.

    Arguments:
       p - ratio of planet radius to stellar radius
       z - distance between star and planet divided by stellar radius
    Returns:
       FluxRatio - ratio of obscured to unobscured stellar flux

    """
    lambd = lam_e(p, np.abs(z))
    assert lambd != -999, 'lam_e failed'
    
    return 1 - lambd
```

Now loop over the z values and compute the predicted flux.

```python
def predicted_lc(z, flux_calc ):
    flux_predicted = []
    for zval in z:
        flux_predicted.append(flux_calc(p, zval))
    return flux_predicted

flux_predicted = predicted_lc(z, FluxRatio)
```

Finally, we can plot it.

```python
def model_obs_compare(time_obs, flux_normalized, flux_normalized_err, flux_pred):
    """plot comparison between model and observation"""
    plot.plot(time_obs, flux_pred, color = "orange")   #keeping data blue for consistency...
    plot_data(time_obs, flux_normalized, flux_normalized_err, ylabl = "Flux (normalized)")

model_obs_compare(first['t'], fnorm['fn'], fnorm['fen'], flux_predicted)
```

well, that doesn't look so great!  Let's refine in two ways: first, make the planet smaller, and second, substantially shorten $\tau$.  (And iterate in this manner, using the next cell, until I reach a good result.)

```python
p = 0.084; tau = 0.093; t0 = 261.297
z = z_from_time(first['t'], t0, tau)

flux_predicted = predicted_lc(z, FluxRatio) 

model_obs_compare(first['t'], fnorm['fn'], fnorm['fen'], flux_predicted)
```

This looks okay -- clearly we could do better, but we'll leave that to later.  The final result I got is $\tau=0.093, t_0 = 261.297$, and  $p = 0.084$.

Use a more realistic limb-darkening function to better reproduce the shape of the eclipse, and combine with the limb darkening integration routine
from Problem Set 2

```python
def simpson_int (integrand, lim_0, lim_1, nx):
    """Extended Simpson's rule integration.
        (1/6 f0 + 2/3 f0.5 + 1/3 f1 + 2/3 f1.5 +... + 1/6 fn)
         integrand: the function f(x) to be integrated.
         lim_0 and lim_1: lower and upper integration limits
         nx: the number of integration steps"""

    delx = (lim_1 - lim_0) / float(nx)  #rectangle width
    third = (1.0/3.0)                   #save dividing at every iteration
    
    #first, second, and last points, which aren't part of regular sequence
    f_x0 = integrand(lim_0)
    f_xh = integrand(lim_0 + 0.5*delx)
    f_xn = integrand(lim_1)
    #print('f_x0, f_xh, f_xn', f_x0, f_xh, f_xn) #debug
    result = (1.0/6.0)*(f_x0 + 4.0*f_xh + f_xn) * delx
    
    #print ('limits, delx, first, second, last', lim_0, lim_1, delx, result)  #debug
    
    #all the rest (note indexes are 1..nx-1 for arange(1,nx))
    for i in np.arange(1,int(nx)):
        x_full = lim_0 + i * delx        #the current ordinate
        x_half = lim_0 + (i+0.5) * delx  #the current half ordinate

        f_xf = integrand(x_full)    #using the function provided
        f_xh = integrand(x_half)
        
        #use += operator here (a += b means a = a+b)
        result += third*(f_xf + 2.0*f_xh)*delx  
        
    return result

def physical_limb(r):
    """returns the value of the physical limb darkening function
    (equation 1 of the third problem set)"""
    mu = m.sqrt(1-r*r)   #cos theta
    return 1.0 - (1.0 - mu**1.5)

def delta(p, r, z):
    """returns the value of the delta function (equation 2, PS2),
    protecting against obvious fault."""
    if (r >= z + p) or (r <= z - p):
        return 0
    elif (r + z <= p):
        return 1
    else:  #all other cases
        if (z != 0) and (r!=0): #prevent div by zero errors
            argument = (z*z - p*p + r*r) / (2.0*z*r)
        else:
            print('Either z or r vanished. z =',z,'; and r =',r,'.')
            return -999   #an error condition, to be checked by callers
        return (1.0/m.pi) * m.acos(argument)


def limb_dark_flux(p, z, int_func=simpson_int, limb_func=physical_limb, nx=1000):
    """return the value of the flux, using equation 1, PS3. p & z are the 
    parameters, int_func is the integration routine, limb_func is the 
    chosen limb darkening profile, and nx is the number of integration 
    steps."""
        
    # z should have the absolute value taken
    z = np.fabs(z)
    
    # first integrand for the numerator
    def numerator(r, p = p, z = z):  #default parameters allow passing data
        delt = delta(p, r, z)
        assert delt != -999, "delta calculation failed at r = "+ str(r) + " z = " + str(z)
        return limb_func(r) * (1.0 - delt) * 2.0 * r
    top    = int_func(numerator, 0.0, 1.0, nx)
    
    #then for the denominator
    def denominator (r):
        return limb_func(r) * 2.0 * r
    bottom = int_func(denominator, 0.0, 1.0, nx)
        
    return top/bottom

assert 0.0 <= limb_dark_flux(p, 0.9) <=1.0  #check for unphysicality
```

Now let's go back and see what this looks like for our old parameters.

First, let's define the two alternate root finders: secant method and bisection.

```python
def RootFinderSecant(function, first, second, tolerance):
    """ Find function root to given tolerance given first two guesses first & second """
    for i in range(100):   # do at most 100 iterations
        deltax = (second-first)*function(second)/(function(second)-function(first))
        first = second
        second = second - deltax  # better to limit this so it doesn't go to a crazy value
        if abs(deltax) < tolerance:
            print("iterations:",i)
            return second
    raise ArithmeticError("root finding failed")
```

```python
def RootFinderBisect(function, left, right, tolerance):
    """ Find function root to given tolerance given bracket left and right """
    
    if (function(left)*function(right) >= 0):
        raise ValueError("Root not bracketed")
    i = 0  # let's count the number of iterations (not necessary but interesting)

    #main 
    while (right-left) > tolerance:
        mid = (right+left)/2.0
        i += 1
        if function(mid)*function(right) <= 0.0:
            left = mid
        else:
            right = mid
            
    print("iterations:",i)
    return mid
```

Let's test each of these.

```python
def linear(x):
    """a linear function to test the root finders."""
    m = 1; b = -1
    return m*x + b
# 0 = x-1 => x = 1 is expected answer

tol5 = 1e-5; tol10 = 1e-10       #root accuracy choices
left = 0.0; right = 2.0          #bracket the root
```

```python
RootFinderBisect(linear, left, right, tol5)
```

```python
RootFinderSecant(linear, left, right, tol5)
```

Now we need to define the actual function we want to find the root of.

```python
def Seff(Teff):
    T0 = 5780.0;  a = 8.177e-5; b = 1.706e-9; Seff_sun = 1.014
    Tstar = Teff - T0
    return Seff_sun + a * Tstar + b * Tstar**2
    
def HabitableZone(Teff):
    T0  = 5780.0; d = 0.5 #AU
    Lum = (Teff/T0)**4
    return m.sqrt(Lum/Seff(Teff)) - d
```

And now let's apply both methods.

```python
left = 1.0; right = 1.0e4    #bracket the stellar temperature

RootFinderBisect(HabitableZone, left, right, tol5)
```

```python
RootFinderBisect(HabitableZone, left, right, tol10)
```

```python
RootFinderSecant(HabitableZone, left, right, tol5)
```

```python
RootFinderSecant(HabitableZone, left, right, tol10)
```

As expected, they both get the same answer to an accuracy of 1e-5, but the Secant method does it with many fewer iterations!  Note: any of these answers would have completed the problem.
