```python

def delta (p, z):
    
    if (1 + p) < z:
        result = 0.0
        return result
    
    elif m.fabs(1 - p) < z and z <= (1 + p):
        k0_value = (p**2 + z**2 - 1) / (2.0*p*z) 
        k1_value = (1 - p**2 + z**2) / (2.0*z)
        if -1 <= k0_value <= 1 and -1 <= k1_value <= 1: #to stay within domain of acos (-1, 1)
            k0       = m.acos(k0_value)
            k1       = m.acos(k1_value)
            result   = ((1/m.pi)*(p**2*k0 + k1 - m.sqrt((4*z**2 - (1 + z**2 - p**2)**2)/4.0)))
        return result

    elif z <= (1 - p):
        result = p**2
        return result
    
    elif z <= (p - 1):
        result = 1
        return result
    
    else:
        result = "Incorrect values"
        return result

# Testing function delta works:
# print (delta(0.2, 0.5))
# Output: 0.04000000000000001 function is correct
```

```python
# Part b) Getting code from PS2

def I (r):           # Defining I(r) with no limb- darkening = 1
    return (1 - (1 - (1 - r**2)**0.5))

def delta (p, r, z): # Function for different conditions of delta
    
    if r >= (z + p) or r <= (z - p):
        result = 0.0
        return result
        
    elif (r + z) <= p:
        result = 1
        return result
        
    else:
        k_value = (z**2 - p**2 + r**2) / (2.0*z*r) # Value to evaluate acos
        if -1 <= k_value <= 1:                     # Boundaries of acos to avoid math error
            k = m.acos (k_value)                   # "k" variable substitutes the equation
            result = (1/m.pi) * k
        return result
    
# Testing delta function for different "r":
#for r in np.arange (0, 1.5, 0.5):
#    print (r, delta(0.2, r, z(t_i, 266.25, (266.3 - 266.12))))
# It works!
```

```python
def function_numer (r):                          # numerator function inside the integral
    return (I(r) * (1 - delta(p, r, z(t_i, 266.25, (266.3 - 266.12))) * 2 * r)) # ratio of two integrals with no limb darkening, i.e. I(r) = 1

def function_denom (r):                          # denominator function inside the integral
    return (I(r) * 2.0 * r) 

def trap_integ1 (function_numer, N, lower_b, upper_b):                                # Trapezoidal rule for integration numerator
    delta_x    = (upper_b - lower_b) / float(N)                                       # N is the number of steps
    area_trap1 = delta_x * 0.5 * (function_numer(lower_b) + function_numer(upper_b))  # Area of a trapezoid
    for i in range (0, N):                                                            # Loop to evaluate at different Ns
        area_trap1 = area_trap1 + delta_x * function_numer(i * delta_x) 
    return area_trap1

def trap_integ2 (function_denom, N, lower_b, upper_b):                                # Integral for the denominator
    delta_x    = (upper_b - lower_b) / float(N) 
    area_trap2 = delta_x * 0.5 * (function_denom(lower_b) + function_denom(upper_b))  # Area of a trapezoid
    for i in range (0, N):
        area_trap2 = area_trap2 + delta_x * function_denom(i * delta_x) 
    return area_trap2

# Testing for integrals:
# j = 1 
# while j < 6:
#    print ("f1(x) =", trap_integ1(function_numer, 10**j, 0, 1)) # Result for integral in numerator
#    print ("f2(x) =", trap_integ2(function_denom, 10**j, 0, 1)) # Result for integral in denominator
#    j += 1
# It works!
```

```python
def function_t (trap_integ1, trap_integ2):   # Function to oftain the ratio of integrals
    transit_f = trap_integ1 / trap_integ2    # F(p,z) = ratio of 2 integrals
    return transit_f

j = 1
while j < 6:
    flux_limb = (function_t (trap_integ1 (function_numer, 10**j, 0, 1), (trap_integ2 (function_denom, 10**j, 0, 1 ))))
    print ("F(p, |z|) =", flux_limb)
    j += 1


# Its not clear if we need to plot the integral as well or not

plt.plot(z_t, flux_limb)
plt.xlabel('Distance $z$')
plt.ylabel('Flux $F$')
plt.show()
```

```python
# Problem 3: Habitable Zone around a star

import numpy as np

# Values given by the problem:

a = 8.177**-5
b = 1.706**-9
S_eff0 = 1.014

# Inner edge of the habitable zone: 
def d_eff (T_eff):
    return ((T_eff / 5780.0)**4 / (S_eff0 + a*(T_eff - 5780) + b*(T_eff - 5780)**2))**0.5  #Surface equation

# Using Bisection Method:
def bisect (d_eff, b1, b2, tol): # Where b1 and b2 are the integrals
    while (b2 - b1) >= tol:
        m = (b1 + b2) / 2.0      # Midpoint
        if d_eff(m) < tol:
            return m
        elif np.sign(d_eff(b1)) == np.sign (d_eff(m)):
            b1 = m
            return m
        else:
            b2 = m
            return m

root = bisect (d_eff, 0, 10**6, 0.1)
print ("The root is:", root)

print ("d(T_eff) =", d_eff (root))

# Determining value of T_eff for radius 0.5 AU
result = d_eff(root) - 0.5
print (result)
```
